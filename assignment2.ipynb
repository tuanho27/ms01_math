# Math for Machine Learning Exercise

```python
from scipy.optimize import linprog
```

```python
obj = [25, 30, 0, 0]
lhs_eq = [[20, 30, 1, 0],
          [5, 4, 1, 0]]
rhs_eq = [60,
          12]
```

```python
 bnd = [(0, float("inf")),  # Bounds of t
        (0, float("inf")),  # Bounds of b
        (0, float("inf")),  # Bounds of y1
        (0, float("inf"))]  # Bounds of y2
    
```

```python
opt = linprog(c=obj,
           A_eq=lhs_eq, b_eq=rhs_eq, bounds=bnd,
           method="revised simplex")
```
